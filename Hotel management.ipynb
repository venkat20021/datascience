```python
import mysql.connector
import tkinter as tk
from tkinter import messagebox


db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="hotel"
)


cursor = db.cursor()

# Function to make a reservation
def make_reservation():
    guest_name = guest_name_entry.get()
    room_number = int(room_number_entry.get())
    check_in_date = check_in_date_entry.get()
    check_out_date = check_out_date_entry.get()

    
    insert_query = "INSERT INTO reservations (guest_name, room_number, check_in_date, check_out_date) VALUES (%s, %s, %s, %s)"
    data = (guest_name, room_number, check_in_date, check_out_date)
    cursor.execute(insert_query, data)
    db.commit()
    messagebox.showinfo("Success", "Reservation made successfully!")


def check_availability():
    check_date = check_date_entry.get()
    room_number = int(room_number_entry2.get())

    query = "SELECT * FROM reservations WHERE room_number = %s AND (%s BETWEEN check_in_date AND check_out_date OR %s BETWEEN check_in_date AND check_out_date)"
    data = (room_number, check_date, check_date)
    cursor.execute(query, data)
    reservations = cursor.fetchall()

    if reservations:
        messagebox.showinfo("Availability", "Room is not available on this date.")
    else:
        messagebox.showinfo("Availability", "Room is available on this date.")


root = tk.Tk()
root.title("Hotel Reservation System")


tk.Label(root, text="Guest Name:").grid(row=0, column=0)
guest_name_entry = tk.Entry(root)
guest_name_entry.grid(row=0, column=1)

tk.Label(root, text="Room Number:").grid(row=1, column=0)
room_number_entry = tk.Entry(root)
room_number_entry.grid(row=1, column=1)

tk.Label(root, text="Check-in Date (YYYY-MM-DD):").grid(row=2, column=0)
check_in_date_entry = tk.Entry(root)
check_in_date_entry.grid(row=2, column=1)

tk.Label(root, text="Check-out Date (YYYY-MM-DD):").grid(row=3, column=0)
check_out_date_entry = tk.Entry(root)
check_out_date_entry.grid(row=3, column=1)


make_reservation_button = tk.Button(root, text="Make Reservation", command=make_reservation)
make_reservation_button.grid(row=4, column=0, columnspan=2)


tk.Label(root, text="Date to Check Availability (YYYY-MM-DD):").grid(row=5, column=0)
check_date_entry = tk.Entry(root)
check_date_entry.grid(row=5, column=1)

tk.Label(root, text="Room Number:").grid(row=6, column=0)
room_number_entry2 = tk.Entry(root)
room_number_entry2.grid(row=6, column=1)


check_availability_button = tk.Button(root, text="Check Availability", command=check_availability)
check_availability_button.grid(row=7, column=0, columnspan=2)


exit_button = tk.Button(root, text="Exit", command=root.destroy)
exit_button.grid(row=8, column=0, columnspan=2)


root.mainloop()


db.close()
```
